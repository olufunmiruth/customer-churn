```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

# Data preprocessing

```python
#check data info
df.info()
```

There are no null values. since we have categorical variables, we need to encode them as our model cannot work well with categorical data.
pd.get_dummies() is used here.Dummies have either 1 or 0 for each row or observation.It produces a new column for each unique observation and it will place one for each column corresponding to the categorical value in that row or obeservation.

```python
#split data into feature and target, also drop customer_id because it is irrelevant
X = pd.get_dummies(df.drop(['Churn','Customer ID'], axis = 1))
```

```python
#the target variable 
y = df.Churn.apply(lambda x : 1 if x == 'Yes' else 0)
```

```python
#split into train and test set
X_train,X_test,y_train, y_test = train_test_split(X,y,test_size = 0.2)
```

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
```

```python
#build model
model = Sequential() ##instantiates the sequential class
model.add(Dense(units = 32,activation ='relu',input_dim =len(X_train.columns)))
model.add(Dense(units = 64,activation = 'relu'))
model.add(Dense(units =1,activation = 'sigmoid'))
```

```python
#compile model
model.compile(loss = 'binary_crossentropy', optimizer = 'sgd',metrics ='accuracy')
```

```python
#fit model
model.fit(X_train, y_train,epochs=500,batch_size = 32)
```

```python
#prediction
y_pred= model.predict(X_test)
```

### the results are continuous values, set values less than 0.5 as 0 and more than 0.5 as 1 since we expect a binary outcome


```python
y_pred = [0 if outcome < 0.5 else 1 for outcome in y_pred]
```

```python
y_pred
```

```python
#evaluate accuracy 
accuracy_score(y_test,y_pred)
```

```python
#save model
model.save('Churn_model')
```
